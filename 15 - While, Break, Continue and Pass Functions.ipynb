WHILE LOOP IN PYTHON:
- Learn about while
- Syntax : 
           while some_boolean_condition:
                # do something
           else:
                # do something else
                
- We have the 3 statements of while : break, continue and pass


```python
#1 This gives an infinite loop of x as the condition never gets false. Hence we interrupt it 

x = 0
while x < 5:
    print (x)
```

```python
#2
x = 0
while x < 5:
    print (x)
    x = x + 1
```

```python
#3
x = 50
while x < 5:
    print (x)
    x += 1
else:
    print("This is not true")
```

```python
#4 Using the pass

x = [1,2,3]
for item in x:
    print(item)
```

```python
x = [1,2,3]
for item in x:
    pass
print('End of the file')
```

```python
#5 Using Continue method

mystring = 'Sammy'
for letter in mystring:
    print(letter)
```

```python
for letter in mystring:
    if letter =='a':           #This omits the letter a - as we parse by it
        continue
    print(letter)
```

```python
for letter in mystring:
    if letter =='a':           #This does not print the letters after a - as we break after it
        break
    print(letter)
```

```python
x = 0
while x < 5:
    if x == 3:
        break
    print(x)
    x += 1
    
```
